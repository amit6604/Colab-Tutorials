# Upload Files & Data 

1) How to upload any file

2) How to manipulate data


```
# Upload csv file

from google.colab import files

file = files.upload()

```

```
# check uploaded file

print(file)
```

```
## Using 'AAPL.csv' as Dictionary key in above file

file = file['AAPL.csv']
```

```
print(file) # All values for Dictionary have been selected.
```

```
# Check rows in file

for item in file:
  print(item)
  
```

```
# Output will be number as datatype is bytes

print(type(file))
```

```
# Convert datatype to string

file = str(file)

print(type(file))
```

```
#Split row by '\n'

file = file.split("\\n")

print(file) # List of strings
```

```
for item in file:
  print(item)
```

```
# Convert each row list of a list

new_list = []

for item in file:
  new_list.append(item.split(','))
  
  
print(new_list)  # each row would be list of list
```

```
for item in new_list:
  print(item)
```
